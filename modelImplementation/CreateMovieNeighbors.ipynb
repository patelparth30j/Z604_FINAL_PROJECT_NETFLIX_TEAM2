```python
import pandas as pd
import numpy as np
```

```python
df_movie = pd.read_csv('movie_features_df.csv', sep = '\t')
df_movie=df_movie.rename(columns = {'Unnamed: 0':'MovieID'})
```

```python
df_movie2 = df_movie.set_index(['MovieID'])
df_movie2
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors(5)
neigh.fit(df_movie2)
```

```python
var = neigh.kneighbors(df_movie2[df_movie2.index==8192], 6, return_distance=True)
var
```

```python
df_movie['MovieID'][df_movie.index==8]
```

```python
import csv
writer = csv.writer(open("/Users/Parth/Desktop/knn_withDistance.csv", 'w'))

for i in df_movie2.index:
    neighbors_distance = neigh.kneighbors(df_movie2[df_movie2.index==i], 6, return_distance=True)
    distance = neighbors_distance[0]
    neighbors = neighbors_distance[1]
    #print neighbors[0]
    for j in range(1,6):
        writer.writerow([i , df_movie['MovieID'][df_movie.index==neighbors[0][j]].to_string(index=False),distance[0][j]])
        
```
